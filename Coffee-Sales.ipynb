# Importing libiraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
d1=pd.read_csv("index_1.csv")
d2=pd.read_csv("index_2.csv")
```

```python
d1
```

```python
d2
```

```python
df=pd.concat([d1,d2])
df
```

```python
df.info()
```

```python
df['date']=df['date'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
df['datetime']=df['datetime'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
df=df.drop(columns=['card'])
```

```python
df
```

```python
df['time']=df['datetime'].dt.time
```

```python
df
```

```python
df['coffee_name'].value_counts()
```

```python
df['hour']=df['datetime'].dt.hour
```

```python
df
```

```python
hourly_Sales=df['hour'].value_counts().sort_index()
```

## Sales on Day time

```python
plt.figure(figsize=(10,5))
plt.plot(hourly_Sales.index,hourly_Sales.values,marker='o')
plt.xticks(range(24))
plt.xlabel('Hour of the day')
plt.ylabel('Number of sales')
plt.title('Hourly sales distribution')
plt.grid(True)
plt.show()


```

# Product Sales by Day time

```python
Sales_hour=df.groupby(['hour','coffee_name']).size().reset_index(name='count')
fig=px.line(Sales_hour,x='hour',y='count',color='coffee_name',title='Sales by hour and product',markers=True)
fig.update_layout(xaxis=dict(dtick=1,title='Hour of Day'),yaxis_title='Number of Sales')
fig.show()





```

## Extracting month from date

```python
df['sales_month']=df['date'].dt.to_period('M').astype(str)
df
```

##### Summary Sales trending by month

```python
Sales_month=df.groupby(['sales_month'])['money'].sum().reset_index()
Sales_month
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=Sales_month, x='sales_month',y='money')
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
df
```

```python
coffee_sales=df.groupby(['coffee_name'])['money'].sum().reset_index()
coffee_sales.sort_values(by='money',ascending=False)
```

```python
len(coffee_sales)
```

```python
for i in range (0,len(coffee_sales),10):
    plt.figure(figsize=(20,10))
    plt.title('Coffee name Sales')
    plt.xlabel('Coffee name')
    plt.ylabel('Sales')
    x=sns.barplot(x=coffee_sales.iloc[i:i+10]['coffee_name'],y=coffee_sales.iloc[i:i+10]['money'],data=coffee_sales)
    for contanier in x.containers:
        x.bar_label(contanier)
    plt.show()
```

#### selecting top sales coffee name

```python
top9=coffee_sales.nlargest(9,'money')
```

```python
top9
```

```python
plt.figure(figsize=(20,10))
plt.title('Coffee name Sales')
plt.xlabel('Coffee name')
plt.ylabel('Sales')
x=sns.barplot(x='coffee_name',y='money',data=top9)
for contanier in x.containers:
    x.bar_label(contanier)
plt.show()
```
